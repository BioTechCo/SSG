```python
import GEOparse
import pandas as pd
import os
```

```python
gse = GEOparse.get_GEO(geo="GSE89093", destdir="./")
```

```python
# inspect a single sample
cancer_status_list = []
table_data = pd.DataFrame()
i=0
for gsm_name, gsm in gse.gsms.items():
    
    print("Name: ", gsm_name)
    # print("Metadata:",)
    # for key, value in gsm.metadata.items():
    #     print(" - %s : %s" % (key, ", ".join(value)))
    # print ("Table data:",)
    # print (gsm.table.head())

    char_dict = {key: value for item in gsm.metadata['characteristics_ch1'] for key, value in [item.split(': ')]}
    cancer_status_list.append(char_dict['cancer status'])
    sample_table = gse.gsms[gsm_name].table
    # change the column name `VALUE` to the f"{i}"
    sample_table = sample_table.rename(columns={"VALUE": f"{i}"})
    if i==0:
        table_data = pd.concat([table_data, sample_table["ID_REF"]], axis=1)
    # merge the sample table of the colums f"{i}" with the previous sample tables
    table_data = pd.concat([table_data, sample_table[f"{i}"]], axis=1)
    i+=1
```

```python
# cancer_status = pd.concat([pd.Series(["dummy"]), pd.Series(cancer_status_list)], ignore_index=True)
cancer_status = pd.DataFrame(cancer_status_list, columns=["cancer_status"])
# insert a row of dummy data to the top of cancer_status
cancer_status = pd.concat([pd.DataFrame(["dummy"], columns=["cancer_status"]), cancer_status], ignore_index=True)
cancer_status
```

```python
# addasd = table_data.T
table_data1 = pd.concat([table_data.T.reset_index(drop=True), cancer_status], ignore_index=True, axis=1)
table_data1
```

```python
table_data1.T.to_csv("GSE89093.csv", index=False)
```

```python
# scan current directory to open the first directory
idat_folder = "idats"
```

```python
file_mapping = {}

for f in os.listdir(idat_folder):
    if f.endswith('.idat'):
        print("idat file found: ", f)
        file_mapping[f.split('_')[0]] = f
        
```

```python
sample_sheet = pd.DataFrame()

characteristics_identifier = "disease state"

for _, gsm in gse.gsms.items():
    gsm_name = gsm.name
    basket = {}
    basket['Sample_Name'] = ''
    basket['Sample_Plate'] = ''
    basket['Pool_ID'] = ''
    char_dict = {key: value for item in gsm.metadata['characteristics_ch1'] for key, value in [item.split(': ')]}
    basket['Sample_Group'] = char_dict[characteristics_identifier]
    if gsm_name in file_mapping:
        print(f"Found {gsm_name} in selected idat files")
        basket['Sentrix_ID'] = file_mapping[gsm_name].split('_')[1]
        basket['Sentrix_Position'] = file_mapping[gsm_name].split('_')[2]
        sample_sheet = pd.concat([sample_sheet, pd.DataFrame([basket])], ignore_index=True)
        
print(sample_sheet)
```

```python
sample_sheet.sort_values(by=['Sample_Group', 'Sentrix_ID', 'Sentrix_Position'], inplace=True)
sample_sheet.reset_index(drop=True, inplace=True)
sample_sheet['Sample_Name'] = sample_sheet.index + 1
sample_sheet.to_csv("sample_sheet.csv", index=False)
```
