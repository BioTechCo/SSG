```python
import GEOparse
import pandas as pd
```

```python
gse = GEOparse.get_GEO(geo="GSE89093", destdir="./")
```

```python
cancer_status_list = []
table_data = pd.DataFrame()
i = 0
for gsm_name, gsm in gse.gsms.items():
    print("Name: ", gsm_name)
    char_dict = {key: value for item in gsm.metadata['characteristics_ch1'] for key, value in [item.split(': ')]}
    cancer_status_list.append(char_dict['cancer status'])
    sample_table = gse.gsms[gsm_name].table
    sample_table = sample_table.rename(columns={"VALUE": f"{i}"})
    if i == 0:
        table_data = pd.concat([table_data, sample_table["ID_REF"]], axis=1)
    table_data = pd.concat([table_data, sample_table[f"{i}"]], axis=1)
    i += 1
```

```python
table_data.to_csv("GSE89093.csv", index=False)
```
